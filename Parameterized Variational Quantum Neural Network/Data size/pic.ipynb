```python
import matplotlib.pyplot as plt
import pandas as pd

# Data from the table
data = {
    "Data Size": [4, 8, 12, 16, 20, 40, 100, 200, 400, 800, 1600, 4000, 8000, 16000],
    "MSE": [0.004858, 0.001168, 0.000956, 0.000627, 0.000473, 0.000571, 0.000558, 0.000507, 0.000452, 0.000391, 0.000373, 0.000362, 0.000347, 0.00035],
    "R^2": [0.638954, 0.913144, 0.928929, 0.953397, 0.964838, 0.957552, 0.95846, 0.96228, 0.96637, 0.97091, 0.97221, 0.97306, 0.97419, 0.97397]
}

df = pd.DataFrame(data)

# Creating the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting MSE with ax1 and setting its color to blue
color = 'tab:blue'
ax1.set_xlabel('Data Size')
ax1.set_ylabel('MSE', color=color)
ax1.plot(df["Data Size"], df["MSE"], color=color, marker='o', label='MSE')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xscale('log')  # Logarithmic scale for better visualization
ax1.set_xticks(df["Data Size"])
ax1.set_xticklabels(df["Data Size"], rotation=45)

# Creating a second y-axis for R^2
ax2 = ax1.twinx()  
color = 'tab:red'
ax2.set_ylabel('$R^2$', color=color)
ax2.plot(df["Data Size"], df["R^2"], color=color, marker='x', label='$R^2$')
ax2.tick_params(axis='y', labelcolor=color)
# Adding titles and grid
plt.title('Out-of-Sample PVQNN Performance on Different Training Data Sizes')
plt.grid(visible=True, which='both', color='0.95')
# Getting handles and labels for both axes
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()

# Combining legends
ax1.legend(handles1 + handles2, labels1 + labels2, loc='upper right',borderaxespad=5)
# Show plot
fig.tight_layout() 
plt.show()
```
